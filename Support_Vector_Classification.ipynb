## **SUPPORT VECTOR CLASSIFICATION**

LOAD LIBRARIES

```python
import pandas as pd
import numpy as np
```

```python
#Load Dataset
from sklearn.datasets import load_breast_cancer
ds=load_breast_cancer()
```

```python
ds.data
```

```python
ds.feature_names
```

```python
#target
ds.target_names
```

```python
#create dataframe
df=pd.DataFrame(np.c_[ds.data,ds.target],columns=[list(ds.feature_names)+['target']])
df.head()
```

```python
df.shape
```

```python
df.tail()
```

```python
df.info()
```

```python
df.describe()
```

```python
#Split data
X=df.drop('target',axis=1)
y=df['target']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=51)
```

```python
print("Shape of X_train",X_train.shape)
print('Shape of X_test',X_test.shape)
print("Shape of y_train",y_train.shape)
print("Shape of y_test",y_test.shape)
```

# **TRAIN SUPPORT VECTOR CLASSIFICATION MODEL

```python
from sklearn.svm import SVC
classifier_rbf=SVC(kernel="rbf")
classifier_rbf.fit(X_train,y_train)
```

```python
classifier_rbf.score(X_test,y_test)
```

FEATURE SCALING

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
sc.fit(X_train)
```

```python
X_train_sc=sc.transform(X_train)
X_test_sc=sc.transform(X_test)
```

```python
#after feature scaling we need to train the data
classifier_rbf2=SVC(kernel='rbf')
classifier_rbf2.fit(X_train_sc,y_train)
```

```python
classifier_rbf2.score(X_test_sc,y_test)
```

**SVC With Kernel Linear**

```python
classifier_linear=SVC(kernel='linear')
classifier_linear.fit(X_train_sc,y_train)
classifier_linear.score(X_test_sc,y_test)
```

**SVC With Kernel Poly**

```python
classifier_poly=SVC(kernel='poly')
classifier_poly.fit(X_train_sc,y_train)
classifier_poly.score(X_test_sc,y_test)
```
